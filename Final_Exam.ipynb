国別に比較的に国民が寛容な国は何をもって国民がその余裕を持てているの調べてみたいと思った

```python
import pandas as pd

df = pd.read_csv("world_happiness_report.csv")

# 基本情報
df.head()

df.info()
df.describe()

import matplotlib.pyplot as plt
from matplotlib import font_manager, rcParams

rcParams["font.family"] = "Yu Gothic"

rcParams["axes.unicode_minus"] = False

```

寛容さの分布を確認


```python
import matplotlib.pyplot as plt

plt.hist(df["Generosity"], bins=30)
plt.xlabel("寛容さ（Generosity）")
plt.ylabel("国の数")
plt.title("寛容さの分布")
plt.show()
```

```python
top_generous = df.sort_values("Generosity", ascending=False).head(10)
top_generous[["Country", "Region", "Generosity", "Happiness Score"]]
```

多くの国で寛容さは低い値を示していて、一部の国が高い値を示す形になっている。

寛容さは何によって生まれる？

```python
plt.scatter(df["Generosity"], df["Happiness Score"])
plt.xlabel("寛容さ（Generosity）")
plt.ylabel("幸福度スコア")
plt.title("寛容さと幸福度スコアの関係")
plt.show()
```

点は全体として右上がり方向に分布していて
寛容さが高い国ほど幸福度がやや高い傾向があると言える。
ただこれが直接的に関係があるかというとそこまで強固には関係がない
弱〜中程度の正の相関が認められる程度

```python
plt.scatter(df["Economy (GDP per Capita)"], df["Generosity"])
plt.xlabel("経済力（1人当たりGDP）")
plt.ylabel("寛容さ（Generosity）")
plt.title("経済力と寛容さの関係")
plt.show()
```

次に経済に関係しているか調べるとまたこれも違う
点は横方向に広く分散しており、明確な右上がり・右下がりの傾向は見られない
経済力が高くなっても、寛容さが一貫して高くなるわけではないことが分かった。

```python
plt.scatter(df["Health (Life Expectancy)"], df["Generosity"])
plt.xlabel("健康（平均寿命）")
plt.ylabel("寛容さ（Generosity）")
plt.title("健康と寛容さの関係")
plt.show()
```

健康寿命との関係性を見てみると点は全体として やや右上がり に分布
経済力×寛容よりも、わずかに関係性が見えやすい。
また健康水準の低い国の方が寛容さも低い値に集中していることも直観に即している。
健康水準と寛容さの間には、弱い正の関係が存在する読み取れる。

```python
plt.scatter(df["Freedom"], df["Generosity"])
plt.xlabel("人生の選択の自由（Freedom）")
plt.ylabel("寛容さ（Generosity）")
plt.title("人生の選択の自由と寛容さの関係")
plt.show()
```

これまでと比較してはっきりとした右上がりの相関がみられる、自己決定の自由がある人間ほど他社に対しても寛容になるのだろうか？

分析の結果、寛容さは経済力とは明確な関係を示さなかった一方で、幸福度や健康、人生の選択の自由といった生活の安定や社会的環境を表す指標とは弱い正の関係が確認された。
特に自由度の高い国では寛容さも高い傾向が見られる

本分析を通じて、経済的および精神的に余裕があり、個人の自己決定が尊重される社会ほど、他者への自発的な支援行動が促進される可能性があるのではないかと考えられる。
